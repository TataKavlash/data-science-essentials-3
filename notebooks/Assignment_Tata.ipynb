```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
income_tn_state = income_tn_deletedcolumns.loc[income_tn_deletedcolumns['county'] == 'Tennessee']
```

```python
income_tn_state.head()
```

```python
plt.figure(figsize = (8, 6))
sns.set(style="whitegrid")
sns.barplot('return_count', 'income_bucket', data = income_tn_state, palette = "Spectral")
plt.xlabel('')
plt.ylabel('')
plt.title('Total Returns per income bucket in Tennessee');
```

```python
income_county_agg['approx_avg_income']= 1000 * income_county_agg['total_inc_amt']/income_county_agg['return_count']
```

```python
income_county_agg.head()
```

```python
data = income_county_agg.sort_values('approx_avg_income') #sort values will sort so lowest amount will be on the left
plt.figure(figsize = (20,10))
plt.bar('county', 'approx_avg_income', data = data, color = "teal")
plt.xticks(rotation = 90, fontsize = 15)
plt.grid(False)
plt.title('Average Income in TN Counties', fontsize =30)
plt.ylabel('Average Income', fontsize = 25);
```
